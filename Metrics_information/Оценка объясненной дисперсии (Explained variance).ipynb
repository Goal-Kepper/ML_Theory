## Оценка объясненной дисперсии (explained_variance_score)

Оценка объясненной дисперсии в рамках машинного обучения является метрикой, которая показывает, насколько хорошо модель объясняет изменчивость данных. Она позволяет оценить, какую часть дисперсии (силы отклонения элементов от среднего значения) в данных можно объяснить с помощью модели.

Значения этой метрики могут принимать значения от 0 до 1, где 0 означает, что модель не объясняет никакой дисперсии в данных, а 1 означает, что модель объясняет всю дисперсию в данных.

Если значение оценки объясненной дисперсии близко к 1, то можно сделать вывод, что модель хорошо объясняет изменчивость данных и может быть использована для прогнозирования новых значений. Если же значение близко к 0, то модель не объясняет изменчивость данных и нуждается в улучшении.

```python
def get_explained_variance_score(y_true, y_pred):
    # вычисляем среднее значение истинных значений
    y_mean_true = sum(y_true) / len(y_true)
    
    # вычисляем среднее значение разности истинных значений и предсказанных
    y_mean_pred = sum(y_true[i] - y_pred[i] for i in range(len(y_true))) / len(y_true)

    # вычисляем среднее значение квадрата отклонения разности истинных и предсказанных значений от их среднего
    ss_true = sum(((y_true[i] - y_mean_true) ** 2 for i in range(len(y_true)))) / len(y_true)

    # вычисляем среднее значение квадрата отклонения разности истинных и предсказанных значений от их среднего
    ss_pred = sum(((y_true[i] - y_pred[i] - y_mean_pred) ** 2 for i in range(len(y_true)))) / len(y_true)

    # вычисляем оценку объясненной дисперсии
    if ss_pred == 0:
      return 1.0
    elif ss_true == 0:
      return 0.0
    else:
      explained_variance = 1 - ss_pred / ss_true
    
    return explained_variance
```

### Сравним реализацию со встроенной функцией в scikit learn

```python
from sklearn.metrics import explained_variance_score

# истинные значения
y_true = [3, -0.5, 2, 7]
# предсказанные значения
y_pred = [2.5, 0.0, 2, 8]

# расчет оценки объясненной дисперсии
explained_variance = explained_variance_score(y_true, y_pred)
my_explained_variance = get_explained_variance_score(y_true, y_pred)

print("Полученная оценка объясненной дисперсии на тесте 1:", explained_variance)
print("Рассчитанная оценка объясненной дисперсии на тесте 1:", my_explained_variance)
```

### Рассмотрим поведение метрики на разных выборках
